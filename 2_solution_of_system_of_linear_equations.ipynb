```python
import numpy as np
import pandas as pd
def Gauss_method(A, b):
    print("Точный метод Гаусса")
    print("A\n", A, "\nb\n", b)
    arr = np.copy(A)
    brr = np.copy(b)
    for k in range(arr.shape[0] - 1):
        #сначала ищем максимальный элемент в столбце
        max_elem = 0
        str = 0
        for i in range (k, arr.shape[0]):
            if abs(arr[i,k]) > abs(max_elem):
                max_elem = arr[i,k]
                str = i
        #меняем строки местами
        change = np.repeat(arr[k], 1)
        arr[k], arr[str] = arr[str], change
        
        change = np.repeat(brr[k], 1)
        brr[k], brr[str] = brr[str], change
        #делим главный элемент на первый коэфициент
        arr[k] = arr[k] / max_elem
        brr[k] = brr[k] / max_elem
        #проходимся по всем строкам, вычитая строки, умноженные на новый коэфициент
        for i in range (k + 1, arr.shape[0]):
            factor = arr[i,k]
            arr[i] = arr[i] - arr[k] * factor
            brr[i] = brr[i] - brr[k] * factor

    #находим аргументы уравнений
    arg = [brr[brr.shape[0] - 1] / (arr[arr.shape[0] - 1, arr.shape[0] - 1])]
    for i in range(arr.shape[0] - 2, -1, -1):
        n = brr[i]
        for j in range(len(arg)):
            n = n - arg[j] * arr[i, arr.shape[0] - 1 - j]
        arg.append(n)
    arg = arg[::-1]
    print("new A\n", arr, "\nnew b\n", brr, "\nsolution\n", np.round(arg))
    
def Gauss_Seidel_method(A, b, eps, res):
    print("Метод простых итераций Зейделя")
    print("A\n", A, "\nb\n", b)
    m = len(A)
    x = [.0 for i in range(m)]
    res["x0"] = x
    converge = False
    pogr = 1.
    k = 1
    iter = 0
    while not converge:
        x_new = np.copy(x)
        for i in range(m):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, m))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        pogr = sum(abs(x_new[i] - x[i])  for i in range(m))
        iter += 1
        x = x_new
        res["x" + str(iter)] = x_new
        converge = pogr < eps
    print("solution\n", np.round(x))
```

```python
A = np.array([[3, 1, 1], [1, 5, 1], [1, 1, 7]])
b = np.array([5, 7, 9])
Gauss_method(A, b)
print()
eps = 1e-6
res = pd.DataFrame()
Gauss_Seidel_method(A, b, eps, res)
res
```
